```python
import pandas as pd
import numpy as np
```

# Opening dat files

# book history

```python
book_history = pd.read_csv('book_history.dat', header=None, sep='\t', engine='python') 
```

```python
#renaming header
good_book_history=book_history.rename(columns={0:'user',1:'item',2:'accessed'})
```

```python
# header is duplicated in a row 0, deleting this row
very_good_book_history=good_book_history.drop([0])
very_good_book_history.head()
```

```python
#converting column item (is string) to numbers
very_good_book_history.item = pd.to_numeric(very_good_book_history.item, errors='coerce')
```

```python
#checking is. item colum is numeric
very_good_book_history.loc[very_good_book_history['item'] == 152]
```

# Joing data frames

```python
#joing data frames
history_and_rating=very_good_book_history.merge(very_good_book_ratings, left_on='user', right_on='user')
```

```python
history_and_rating.head()
```

```python
history_and_rating['user'].value_counts()
#there are 1295 users
```

```python
history_and_rating['item_x'].value_counts()
# there 17154 books
```

```python
history_rating_userinfo=history_and_rating.merge(very_good_users_info, left_on='item_x', right_on='User-ID')
```

```python
history_rating_userinfo.head()
```

```python
history_rating_userinfo_iteminfo=history_rating_userinfo.merge(items_info, left_on='item_x', right_on='book_id')
```

```python
history_rating_userinfo_iteminfo.head()
```

```python
history_rating_userinfo_iteminfo['item_x'].max()
```

```python
books_data=history_rating_userinfo_iteminfo.drop(columns=['item_x', 'smth','item_y','user'])
```

```python
books_data.head()
```

```python
books_data =books_data[['User-ID', 'age', 'location','accessed', 'book_id', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'ISBN', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']]
```
