```python
import pandas as pd
from collections import Counter
import random
import operator
```

```python
#reading data
data=pd.read_csv('../data/books_data.csv')
```

```python
#droping unnecessary column
data.drop(['Unnamed: 0'],axis=1, inplace=True)
```

```python
data.head()
```

# Analysis

```python
#investigating user column
data['age'].value_counts()
```

```python
data.age.unique()
#there are some cells in age column with unrealistic values such as 148
```

```python
# as there are many lines with unrealistic age values, they would be replaced with a mean value
data['age'].astype(float).mean()
#mean age is 36
```

```python
# function to replace unrealistic age
def unreal_age (row):
    if row['age'] > 100 :
        return 36
    elif row['age'] < 10:
        return 36
    else:
        return row['age']
    
age_list=data.apply (lambda row: unreal_age(row), axis=1)
data['Client age'] = age_list
```

```python
data.head()
```

```python
data['Client age'].unique()
```

```python
# function to create age groups
def age_groups (row):
    if row['Client age'] <= 16 :
        return 'teenagers'
    elif row['Client age'] >16 and row['Client age']<30 :
        return 'youth and students'
    elif row['Client age'] >=30 and row['Client age'] <55 :
        return 'adults'
    else:
        return 'pensioner'
    
age_group_list=data.apply (lambda row: age_groups(row), axis=1)
data['age group'] = age_group_list
```

```python
data['age group'].value_counts()
#main customer group is adults
```

```python
#creatind dataframes by age group
teenagers=data[data['age group']=='teenagers']
adults=data[data['age group']=='adults']
youth=data[data['age group']=='youth and students']
pensioner=data[data['age group']=='pensioner']

#creating list of books of each dataframes by age group
teenagers_list = teenagers['Book-Title'].tolist()
adults_list = adults['Book-Title'].tolist()
youth_list = youth['Book-Title'].tolist()
pensioner_list = pensioner['Book-Title'].tolist()
```

```python
# creating top 10 list of teen books
teen_list1=Counter(teenagers_list).keys() # equals to list(set(words))
teen_list2=Counter(teenagers_list).values()
top_teen = dict(zip(teen_list1, teen_list2))

sorted_top_teen = dict(sorted(top_teen.items(),
                            key=operator.itemgetter(1),
                            reverse=True))

teen_recomends=[]

for x in list(sorted_top_teen)[0:10]:
    teen_recomends.append(x)
```

```python
# creating top 10 list of adults books
adults_list1=Counter(adults_list).keys() # equals to list(set(words))
adults_list2=Counter(adults_list).values()
top_adults = dict(zip(adults_list1, adults_list2))

sorted_top_adults = dict(sorted(top_adults.items(),
                            key=operator.itemgetter(1),
                            reverse=True))

adults_recomends=[]

for x in list(sorted_top_adults)[0:10]:
    adults_recomends.append(x)
```

```python
# creating top 10 list of youth books
youth_list1=Counter(youth_list).keys() # equals to list(set(words))
youth_list2=Counter(youth_list).values()
top_youth = dict(zip(youth_list1, youth_list2))

sorted_top_youth = dict(sorted(top_youth.items(),
                            key=operator.itemgetter(1),
                            reverse=True))

youth_recomends=[]

for x in list(sorted_top_youth)[0:10]:
    youth_recomends.append(x)
```

```python
# creating top 10 list of pensioner books
pensioner_list1=Counter(pensioner_list).keys() # equals to list(set(words))
pensioner_list2=Counter(pensioner_list).values()
top_pensioner = dict(zip(pensioner_list1, pensioner_list2))

sorted_top_pensioner = dict(sorted(top_pensioner.items(),
                            key=operator.itemgetter(1),
                            reverse=True))

pensioner_recomends=[]

for x in list(sorted_top_pensioner)[0:10]:
    pensioner_recomends.append(x)
```

```python
#creating a small sample for testing
sample=data[data['User-ID']==152]
```

```python
#function that makes recomendation
#checking  function on a small sample
def book_recomends (row):
    for book in sample['Book-Title']:
        if book in teenagers_list:
            return random.choice(teen_recomends)
        elif book in adults_list:
            return random.choice(adults_recomends)
        elif book in youth_list:
            return random.choice(youth_recomends)
        else:
            return random.choice(pensioner_recomends)

recomends_list=sample.apply (lambda row: book_recomends(row), axis=1)
sample['recomends'] = recomends_list
```
